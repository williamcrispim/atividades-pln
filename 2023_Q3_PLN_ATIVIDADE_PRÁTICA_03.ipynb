#### **EXEMPLOS PRÁTICOS**

Vamos importar a classe `WebClient` da biblioteca `slack-sdk`, possibilitando a criação de um cliente para interagir com a API do Slack em Python. Isso habilita a realização de ações programáticas, como envio de mensagens e gerenciamento de recursos no Slack.

### Exemplo 1: Envio de Mensagem para um Canal

Neste exemplo, vamos usar a API do Slack para enviar uma mensagem para um canal específico.

Canal para o qual você deseja enviar a mensagem (substitua pelo nome do seu canal):

```
channel_name = 'atividade-03-pln'
```

Mensagem que você deseja enviar:

```
message_text = 'Esta é uma mensagem de exemplo do Slack API com Python!'
```

Encontre o ID do canal com base no nome:

```
response = client.conversations_list()
channel_id = None
for channel in response['channels']:
    if channel['name'] == channel_name:
        channel_id = channel['id']
        break
```

Envie a mensagem:

```
if channel_id:
    response = client.chat_postMessage(
        channel=channel_id,
        text=message_text
    )
    if response['ok']:
        print(f'Mensagem enviada com sucesso: {response["ts"]}')
    else:
        print(f'Erro ao enviar a mensagem: {response["error"]}')
else:
    print(f'Canal "{channel_name}" não encontrado.')
```

*Lembre-se de substituir `channel_name` pelo nome do canal para o qual deseja enviar a mensagem.*

### Exemplo 2: Listagem de Canais

Neste exemplo, vamos listar todos os canais públicos em um espaço de trabalho Slack.

```
response = client.conversations_list(types='public_channel')

if response['ok']:
    for channel in response['channels']:
        print(f'Nome do Canal: {channel["name"]}, ID do Canal: {channel["id"]}')
else:
    print(f'Erro ao listar os canais públicos: {response["error"]}')
```

### Exemplo 3: Enviar Mensagem Formatada com Anexos

Neste exemplo, você cria uma mensagem com um anexo formatado e a envia para um canal. Isso é útil para exibir informações de maneira mais estruturada.

Crie uma mensagem com anexo

```
message = {
    'channel': '#atividade-03-pln',
    'text': 'Mensagem com anexo:',
    'attachments': [
        {
            'title': 'Teste para a atividade 3',
            'text': 'Texto do anexo com formatação para a atividade da disciplina de PLN.',
            'color': '#36a64f'
        }
    ]
}
```

Envie a mensagem

```
response = client.chat_postMessage(**message)

if response['ok']:
    print(f'Mensagem com anexo enviada com sucesso: {response["ts"]}')
else:
    print(f'Erro ao enviar a mensagem com anexo: {response["error"]}')
```

### Exemplo 4: Carregar e Compartilhar um Arquivo

Você pode usar a API do Slack para carregar e compartilhar arquivos diretamente em canais ou mensagens.

Importar a classe `SlackApiError`, que facilita o tratamento de erros ao interagir com a API do Slack por meio da biblioteca `slack-sdk`.







```
from slack_sdk.errors import SlackApiError
```

Carregue um arquivo

```
try:
    response = client.files_upload(
        channels='#atividade-03-pln',
        file='/content/sample_data/anscombe.json', # Aqui você pode colocar qualquer arquivo que desejar
        title='anscombe.json',
        initial_comment='Arquivo enviado automaticamente.'
    )

    if response['ok']:
        print(f'Arquivo carregado com sucesso: {response["file"]["name"]}')
    else:
        print(f'Erro ao carregar o arquivo: {response["error"]}')
except SlackApiError as e:
    print(f'Erro ao carregar o arquivo: {e}')
```
